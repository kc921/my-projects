```python
df = pd.read_excel("Q3.xlsx",usecols="A,E,H")
print(df)
```

```python
df['受僱人員報酬占GDP比率']=df['受僱人員報酬']/df['國內生產毛額']
print(df)
```

```python
plt.plot(df["統計期"],df["受僱人員報酬占GDP比率"]*100)
plt.xlabel('year')
plt.ylabel('share of compensation of employees to GDP (%)')
plt.title("the share of compensation of employees to GDP from 1981-2020", fontsize = 15, fontweight = "bold", y = 1.1) 
plt.show()
```
